```python
import lxml
import requests
from bs4 import BeautifulSoup

headers = ({'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36'})

rating_map = {1: "One", 2: "Two", 3: "Three", 4: "Four", 5: "Five"}

num = int(input("Enter rating (1 to 5): ").strip())
rating_input = rating_map.get(num)

if rating_input:
    all_books = []
    page = 1
    while True:
        url = f"https://books.toscrape.com/catalogue/page-{page}.html"

        c = requests.get(url, headers=headers)

        soup = BeautifulSoup(c.content, 'lxml')
        book_divs = soup.find_all("article", class_="product_pod")
        if not book_divs:
            break

        for div in book_divs:
            rating_tag = div.find("p", class_=["star-rating"])
            if not rating_tag:
                continue
            classes = rating_tag.get("class", [])
            if rating_input not in classes:
                continue
            title_tag = div.find("h3").find("a")
            price_tag = div.find("p", class_="price_color")
            if not title_tag or not price_tag:
                continue
            title = title_tag.get("title")
            price = price_tag.get_text().strip()
            all_books.append((rating_input, title, price))

        page += 1

    f = open('filtered_books.txt', 'w')
    for rating, title, price in all_books:
        f.write(rating + " | " + title + " | " + price + "\n")
    f.close()

    for rating, title, price in all_books:
        print(rating, "|", title, "|", price)
else:
    print("Invalid input. Please enter a number between 1 and 5.")
```

```python
import lxml
import requests
from bs4 import BeautifulSoup

headers = ({'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36'})

rating_map = {1: "One", 2: "Two", 3: "Three", 4: "Four", 5: "Five"}

num = int(input("Enter rating (1 to 5): ").strip())
rating_input = rating_map.get(num, None)

if rating_input:
    all_books = []
    page = 1
    while True:
        url = f"https://books.toscrape.com/catalogue/page-{page}.html"

        c = requests.get(url, headers=headers)
        if c.status_code != 200:
            break

        soup = BeautifulSoup(c.content, 'lxml')
        book_divs = soup.find_all("article", class_="product_pod")
        if not book_divs:
            break

        for div in book_divs:
            rating_tag = div.find("p", class_=["star-rating"])
            if not rating_tag:
                continue
            classes = rating_tag.get("class", [])
            if rating_input not in classes:
                continue
            title_tag = div.find("h3").find("a")
            price_tag = div.find("p", class_="price_color")
            if not title_tag or not price_tag:
                continue
            title = title_tag.get("title")
            price = price_tag.get_text().strip()
            all_books.append((rating_input, title, price))

        page += 1

    f = open('filtered_books.txt', 'w')
    for rating, title, price in all_books:
        f.write(rating + " | " + title + " | " + price + "\n")
    f.close()

    for rating, title, price in all_books:
        print(rating, "|", title, "|", price)
else:
    print("Invalid input. Please enter a number between 1 and 5.")
```
